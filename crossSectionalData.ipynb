# Loading and Cleaning Cross-Sectional Data

## Loading

```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
import time
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
cg = CoinGeckoAPI()
cg
```

```python
data = cg.get_coins_markets(vs_currency='USD', per_page=250, page=1, order='market_cap_desc')
data
```

```python
df_raw = pd.DataFrame(data)
df_raw
```

```python
page = 2
while True:
    data = cg.get_coins_markets(vs_currency='USD', per_page=250, page=page, order='market_cap_desc')
    if len(data) == 0:
        print('Download complete')
        break
    else:
        df_raw = pd.concat([df_raw, pd.DataFrame(data)], ignore_index=True)
        page += 1
        print('Page {}'.format(page), end='\r')
        #time.sleep(1)
```

```python
df_raw
```

```python
df_raw.info()
```

## Cleaning

```python
import numpy as np
```

```python
df_raw.fillna(np.nan, inplace=True) # replace none by NaN
```

```python
df_raw.drop(columns=['roi'], inplace=True) # remove useless column
```

```python
date_columns = ['ath_date', 'atl_date', 'last_updated']
```

```python
# Right data type for date columns
for column in date_columns:
    df_raw[column] = pd.to_datetime(df_raw[column])
df_raw.info()
```

```python
# Check ids for duplicates
df_raw.id.value_counts()
```

```python
# Check for duplicates
df_raw.name.value_counts()
```

```python
# Duplicates OK here
df_raw.symbol.value_counts()
```

```python
df_raw[df_raw.duplicated(keep=False)] #fully duplicated rows
```

```python
df_raw.drop_duplicates(subset=['id'], keep='first', inplace=True)
df_raw.info()
```

```python
df_raw[df_raw.symbol == 'meta']
```

```python
# Dropping rows unsuitable for analysis
df_raw.dropna(axis=0, subset=['current_price', 'ath', 'atl'], how='all', inplace=True)
df_raw.info()
```

```python
df_raw.loc[(df_raw.market_cap > 0) & (df_raw.circulating_supply == 0)] # Zero supply but market cap > 0
```

```python
df_raw.loc[(df_raw.market_cap > 0) & (df_raw.circulating_supply == 0), 'circulating_supply'] = np.nan
```

```python
df_raw.loc[(df_raw.market_cap > 0) & (df_raw.circulating_supply == 0)] # Zero price but market cap > 0
```

```python
df_raw[(df_raw.current_price < 0) | (df_raw.market_cap < 0)]
```

```python
df_raw = df_raw.loc[~(df_raw.current_price < 0) | (df_raw.market_cap < 0)]
```

```python
df_raw.info()
```

```python
df_raw.describe()
```

```python
df_raw[df_raw.atl < 0]
```

```python
df_raw.sort_values(by='market_cap', ascending=False, inplace=True)
df_raw.head()
```

```python
df_raw['market_cap_rank'] = df_raw.market_cap.rank(method='min', ascending=False)
df_raw.info()
```

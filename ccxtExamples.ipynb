### Working with the since parameter

```python
start = '2020-05-15 00:00:00Z'
```

```python
unix = binance.parse8601(start)
unix
```

```python
# since, but max is 1000 bars
data = binance.fetch_ohlcv(symbol='BTC/USD', since=unix, limit=1000, timeframe='1d')
data
```

```python
len(data)
```

```python
# converting back
binance.iso8601(unix)
```

```python
def get_history(symbol, interval, start=None, limit=1000):
    if start:
        start = binance.parse8601(start)

    data = binance.fetch_ohlcv(symbol=symbol, timeframe=interval, since=start, limit=limit)

    df = pd.DataFrame(data)
    df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
    df.Date = pd.to_datetime(df.Date, unit='ms')
    df.set_index('Date', inplace=True)

    return df
```

```python
df = get_history(symbol='BTC/USD', interval='1h', limit=1000)
df
```

```python
df.info()
```

```python
start = '2021-11-15 09:00:00'
```

```python
df = get_history(symbol='BTC/USD', interval='1h', start=start, limit=1000)
df
```

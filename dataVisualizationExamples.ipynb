# Visualization with Matplotlib

### The plot() Method

```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot()
```

```python
plt.show()
```

```python
titanic.plot(subplots=True, figsize=(15, 12))
plt.show()
```

```python
titanic.plot(subplots=True, figsize=(15, 12), sharex=False)
plt.show()
```

```python
titanic.age.plot(figsize=(12, 8))
plt.show()
```

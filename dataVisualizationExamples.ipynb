### The plot() Method

```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

```python
titanic.age.plot(figsize=(12, 8))
plt.show()
```

#### Customization of Plots

```python
xticks = [x for x in range(0, 901, 50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c='r', linestyle='-', xlim=(0,900), ylim=(0,80), xticks=xticks, yticks=yticks, rot=45)
plt.title('Titanic - Ages', fontsize=15)
plt.legend(loc='best', fontsize=15)
plt.xlabel('Passenger No', fontsize=13)
plt.ylabel('Age', fontsize=13)
plt.grid()
plt.show()
```

### Histograms

```python
plt.style.use('seaborn')
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
titanic.age.value_counts(bins=10)
```

```python
# bin defaluts = 10
# bins, density and cumulative parameters not listed but can be passed
titanic.age.plot(kind='hist', figsize=(12,8), fontsize=15, bins=80)
plt.show()
```

```python
# alternative
titanic.age.hist(figsize=(12,8), bins=80, xlabelsize=15, ylabelsize=15)
plt.show()
```

```python
# Third altenative
# May need to dropna() e.g. titanic.age.dropna()
# density give relative quantity
plt.figure(figsize=(12,8))
plt.hist(titanic.age, bins=80, density=True, cumulative=True)
plt.show()
```

### Scatter Plots

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
titanic.plot(kind='scatter', figsize=(15,8), x='age', y='fare', c='cyan', marker='x', s=50)
plt.show()
```

```python
# Passing a third column to the color parameter give a 3D graph
# colormap changes the bar
titanic.plot(kind='scatter', figsize=(15,8), x='age', y='fare', c='survived', marker='x', s=50, colormap='plasma')
plt.show()
```
